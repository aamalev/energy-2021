```python
def get_m(row):

    b = row.month_year
    if b <2020:
        return 1
    elif b <2020.05:
        return 2
    elif b < 2020.09:
        return 3
    elif b < 2021.01:
        return 4
    elif b < 2021.05:
        return 5
    elif b < 2021.09:
        return 6
    else:
        return 7

data1['kvartal'] = data1.apply(get_m, axis=1)
data1.kvartal.unique()
```

```python
clientl = list(data1.id.unique())
client = pd.DataFrame(index=clientl)

dv = data1[data1.tariff=='двухтарифная'].id
odn = data1[data1.tariff=='однотарифная'].id
client['tariff'] = 3
client['tariff'][dv] = 2
client['tariff'][odn] = 1
```

```python
client['e_kv1'] = 0
client['e_kv2'] = 0
client['e_kv3'] = 0
client['e_kv4'] = 0
client['e_kv5'] = 0
client['e_kv6'] = 0
client['e_kv7'] = 0

client['p_kv1'] = 0
client['p_kv2'] = 0
client['p_kv3'] = 0
client['p_kv4'] = 0
client['p_kv5'] = 0
client['p_kv6'] = 0
client['p_kv7'] = 0
```

```python
# clientl
```

```python
for i in clientl:
    data3 = data1[(data1.id==i)]
    for j in list(range(1,8)):
        data2 = data3[(data3.kvartal==j)]
        client['e_kv'+str(j)][i] = np.mean(data2.used_energy_capacity_kVth)
        client['p_kv'+str(j)][i] = np.mean(data2.used_energy_cost_with_nds_rub)
    
    
    
```

```python
client
```

```python
data2 = data1[data1.tariff=='однотарифная']
data2.dropna(inplace=True)
data2 =data2[(data2.used_energy_capacity_kVth>0)]
data2['price'] = data2.used_energy_cost_with_nds_rub/data2.used_energy_capacity_kVth
np.mean(data2.price)# цена квт однотариф
```

```python
def get_m1(row):

    b = np.mean([row.p_kv1,row.p_kv2,row.p_kv3,row.p_kv4,row.p_kv5,row.p_kv6,row.p_kv7])
    return b

client['al1'] = client.apply(get_m1, axis=1)
```

```python
def get_m2(row):

    b = np.mean([row.e_kv1,row.e_kv2,row.e_kv3,row.e_kv4,row.e_kv5,row.e_kv6,row.e_kv7])
    return b

client['al_en'] = client.apply(get_m2, axis=1)
```

```python
price1 = np.mean(data2.price)
client['al2']=client['al1']
client['al3'] =client['al1']

cl_ind = client[client.tariff!=1].index
client['al1'][cl_ind]= client['al_en'][cl_ind]*price1

    
```

```python
data3 = data1[data1.tariff=='двухтарифная']
data2=data3[data3.day_period=='день']
data2.dropna(inplace=True)
data2 =data2[(data2.used_energy_capacity_kVth>0)]
data2['price'] = data2.used_energy_cost_with_nds_rub/data2.used_energy_capacity_kVth
pice21=np.mean(data2.price)# цена квт однотариф

data2=data3[data3.day_period=='ночь']
data2.dropna(inplace=True)
data2 =data2[(data2.used_energy_capacity_kVth>0)]
data2['price'] = data2.used_energy_cost_with_nds_rub/data2.used_energy_capacity_kVth
pice22=np.mean(data2.price)# цена квт однотариф
print(pice21,pice22)
```

```python
cl_ind = client[client.tariff==3].index
for i in cl_ind:
    data3 = data1[data1.id==i]
    c1 = data3[data3.day_period=='ночь']
    
    c11 = np.mean(c1.used_energy_capacity_kVth)
    
    c2 = data3[data3.day_period!='ночь']
    c22 = np.mean(c2.used_energy_capacity_kVth)
    client['al2'][i] = c22*pice21 +c11*pice22
```

```python
client2=client[client.tariff==2]
listt=['e_kv1', 'e_kv2', 'e_kv3', 'e_kv4', 'e_kv5', 'e_kv6', 'e_kv7']
from sklearn.cluster import KMeans
import numpy as np
kmeans2 = KMeans(n_clusters=10, random_state=11).fit(client2[listt])

client2['clasters']=kmeans2.predict(client2[listt])
client3 = client[client.tariff==1]
client3['clasters'] = kmeans2.predict(client3[listt])
for i in range(10):
    ds = np.mean(client2[client2.clasters==i].al2/client2[client2.clasters==i].al_en)
    ind = client3[client3.clasters==i].index
    client['al2'][ind] =  ds *client['al_en'][ind]
    
```

```python
client2=client[client.tariff==3]
listt=['e_kv1', 'e_kv2', 'e_kv3', 'e_kv4', 'e_kv5', 'e_kv6', 'e_kv7']
from sklearn.cluster import KMeans
import numpy as np
kmeans3 = KMeans(n_clusters=2, random_state=11).fit(client2[listt])

client2['clasters']=kmeans3.predict(client2[listt])
client3 = client[client.tariff!=3]
client3['clasters'] = kmeans2.predict(client3[listt])
for i in range(2):
    ds = np.mean(client2[client2.clasters==i].al3/client2[client2.clasters==i].al_en)
    ind = client3[client3.clasters==i].index
    client['al3'][ind] = ds *client['al_en'][ind]
    
```

```python
def get_m3(row):

    b = 1+np.argmin([row.al1,row.al2,row.al3])
    return b

client['best_tariff'] = client.apply(get_m3, axis=1)
```
